```
import pandas as pd
import numpy as np
```

```
test=pd.read_csv("/content/drive/MyDrive/AUST IDC MEMBER DATABASE (2017 - ) - 1902.csv", dtype=str)
```

```
for c in test.iterrows():
  if c[1][4]=="1st":
    test.replace(to_replace =c[1][4],  
                 value = "1",  
                  inplace = True)
  elif c[1][4]=="2nd":
    test.replace(to_replace =c[1][4],  
                 value = "2",  
                  inplace = True)
  elif c[1][4]=="3rd":
    test.replace(to_replace =c[1][4],  
                 value = "3",  
                  inplace = True)
  elif c[1][4]=="4th":
    test.replace(to_replace =c[1][4],  
                 value = "4",  
                  inplace = True)
```

```
for c in test.iterrows():
  if c[1][5]=="1":
    test.replace(to_replace =c[1][4],  
                 value = c[1][4]+".1",  
                  inplace = True)
  elif c[1][5]=="2":
    test.replace(to_replace =c[1][4],  
                 value = c[1][4]+".2",  
                  inplace = True)
```

```
for name, dtype in test.dtypes.iteritems():
    print(name, dtype)
```

```
cols=["YEAR", "SEMESTER"]
test[cols] = test[cols].apply(lambda s: s.astype(str))
```
